```python
# Function to convert list to string
def list_to_string(input_list, separator=" "):
    if isinstance(input_list, list):
        return separator.join(input_list)
    return input_list  # In case it's already a string, just return it

# Convert the 'genre' and 'synopsis' columns to strings
animes['genre'] = animes['genre'].apply(list_to_string)
animes['synopsis'] = animes['synopsis'].apply(list_to_string)

# Now concatenate 'genre' and 'synopsis' columns
animes['combined'] = animes['synopsis']+ animes['genre']

# Output the result
print(animes[['genre', 'synopsis', 'combined']])
```

```python
import re
def remove_brackets(text):
    return re.sub(r'\[|\]', '', text)

# Apply this function to the 'combined' column
animes['combined'] = animes['combined'].apply(remove_brackets)

animes.head()
```

```python
# Define a function to clean the 'combined' column
def clean_combined_column(text):
    # Add space after "MAL" and remove commas and single quotes
    text = re.sub(r"Written by MAL Rewrite", "Written by MAL Rewrite ", text)
    text = text.replace("'", "")  # Remove single quotes
    text = text.replace(",", "")  # Remove commas
    return text

# Apply this function to the 'combined' column
animes['combined'] = animes['combined'].apply(clean_combined_column)
```

```python
new_animes=animes[['uid','title','combined']]
new_animes.head()
new_animes.iloc[0]['combined']
```

```python
new_animes['combined']=new_animes['combined'].apply(lambda x:x.lower())

new_animes.iloc[0]['combined']
```

```python
#recommendation

#transformation
import nltk
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stems(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)
```

```python
new_animes['combined']=new_animes['combined'].apply(stems) #stemimg:the process of reducing words to their base or root form

new_animes.iloc[0]['combined']
```

```python
new_animes.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')           #most frequent words and stop_words will get rid of unnecessary words??
```

```python
vector=cv.fit_transform(new_animes['combined']).toarray()
vector
```

```python
vector.shape    #rows,columns
```

```python
#to measure the cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(anime):
  index=new_animes[new_animes['title']==anime].index[0]
  distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])    #enumerate generates indexes and identify similarity score

  for i in distances[1:11]:
    print(new_animes.iloc[i[0]].title)
```

```python
recommend('Naruto')
```
